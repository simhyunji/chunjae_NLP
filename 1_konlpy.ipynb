### konlpy 형태소(뜻을 가진 가장 작은 단위) 분석
- 자립형태소: 조사, 어미 없이 자립으로 쓸 수 있음 (명사)
- 의존형태소: 다른 형태소와 결합해서 쓰는 형태소
* morphs : 형태소 추출
* pos : 품사 태깅
* nouns : 명사만 추출

```python
from konlpy.tag import Kkma, Hannanum, Komoran, Okt # Okt -> 트위터 형태

hananum = Hannanum()
kkma = Kkma()
komoran = Komoran()
okt = Okt()
```

형태소 추출 (morphs)

```python
# hannanum : KAIST SWRC(Semantic Web Research Center)에서 개발 -한나눔
hananum.morphs(sentences)
```

```python
# KKma : 서울대학교 IDS(intelligent Data Systems) 연구실에서 개발 -꼬꼬마
kkma.morphs(sentences)
```

```python
# Komoran - 코몬
komoran.morphs(sentences)
```

```python
# okt: 오픈 소스 한국어 분석기, 과거 트위터 형태소 분석기, 최근 버전이라, 간략하게 사용하기 좋음
okt.morphs(sentences)
```

```python
# nsmc 데이터 로드 - 자연어처리할 때 가장 많이 사용되는 데이터
# 'https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt'
train_data = pd.read_csv('https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt', sep='\t', encoding='utf-8')
train_data.to_csv('./data/ratings_train.csv', index=False, encoding='utf-8')
```

```python
# N번째 문서 가져오기
sample_data2 =train_data.iloc[100]['document']
sample_data2 
```

```python
# compare_morphs method 만들기 -> 한 번에 모든 morphs 를 볼 수 있게 출력해보기
def compare_morphs(sample_data2):
    sample_data2 =train_data.iloc[100]['document']
    a= hananum.morphs(sample_data2)
    b= kkma.morphs(sample_data2)
    c= komoran.morphs(sample_data2)
    d= okt.morphs(sample_data2)
    return print('hananum:',a,'\nkkma:',b, '\nkor_text:',c, '\nokt:',d)
```

```python
# 실행
compare_morphs(sample_data2)
```

### 품사, 명사 별 count 하기, 기초 시각화

```python
# train_data에 대해 품사 별 count 하기
# {'Noun': 500, 'Adverb': 44652...}
# okt 활용
# N번째 문서 가져오기
sample_data3 =train_data.iloc[:1000]['document']
str(sample_data3)

# {'Adjective': 189508, 'Adverb': 44652, 'Alpha': 12954, 'Conjunction': 4135, 'Determiner': 17825, 'Eomi': 518, 'Exclamation': 6353, 'Foreign': 6605, 'Hashtag': 4, 'Josa': 360544, 'KoreanParticle': 27531, 'Noun': 896632, 'Number': 27229, 'PreEomi': 14, 'Punctuation': 204877, 'ScreenName': 5, 'Suffix': 46007, 'Unknown': 0, 'Verb': 266856, 'Modifier': 32968, 'VerbPrefix': 10894, 'URL': 48, 'Email': 9}
```

```python
from konlpy.tag import Okt

sample_data3_str = sample_data3.astype(str)

# sample_data3_str을 리스트로 변환하여 각 문자열에 대해 형태소 분석
result = [okt.pos(sentence) for sentence in sample_data3_str.tolist()]
print(result)
```

```python
# 형태소 분석 결과를 담을 딕셔너리 초기화
pos_counts = {
    'Adjective': 0, 'Adverb': 0, 'Alpha': 0, 'Conjunction': 0, 'Determiner': 0,
    'Eomi': 0, 'Exclamation': 0, 'Foreign': 0, 'Hashtag': 0, 'Josa': 0,
    'KoreanParticle': 0, 'Noun': 0, 'Number': 0, 'PreEomi': 0, 'Punctuation': 0,
    'ScreenName': 0, 'Suffix': 0, 'Unknown': 0, 'Verb': 0, 'Modifier': 0,
    'VerbPrefix': 0, 'URL': 0, 'Email': 0
}

# sample_data3_str을 리스트로 변환하여 형태소 분석 후 품사 카운트 업데이트
for sentence in sample_data3_str.tolist():
    pos_result = okt.pos(sentence)
    for word, pos in pos_result:
        if pos in pos_counts:
            pos_counts[pos] += 1
            
# 결과 출력
print(pos_counts)
```

```python
pos_dict = {}
for tag in okt.tagset.keys():
    pos_dict[tag] = 0

for doc in train_data['document'][:1000]:
    doc_pos = okt.pos(str(doc)) # (단어, 품사) 의 튜플 형태 인데 품사 수가 궁금해서 1번 인덱스 가져오기
    for uni_pos in doc_pos:
        pos = uni_pos[1]

        if pos in pos_dict:
            pos_dict[pos] = pos_dict[pos] + 1

        else:
            pos_dict[pos] = 1

print(pos_dict)
```

```python
# plot.bar 그려보기
pd.DataFrame(pos_dict.values(), index=pos_dict.keys(), columns=['counting']).plot.bar()
```
